# Analysis of the Sampling Rate for CHUNKS

Builds on the sampling rates that were generated in the create_sampling_rates_for_chunks.ipynb

## Load Sampling Rate Data for CHUNKS

```python
import pandas as pd
import pyarrow as pa

sampling_rates_for_chunkid = pd.read_parquet('./data/sampling_rates_for_chunkid.parquet', engine='pyarrow')
```

```python
# create subsets for each ITEMID
sampling_rates_for_chunkid_HR = sampling_rates_for_chunkid[sampling_rates_for_chunkid['ITEMID'] == 220045]
sampling_rates_for_chunkid_O2 = sampling_rates_for_chunkid[sampling_rates_for_chunkid['ITEMID'] == 220277]
sampling_rates_for_chunkid_NBP = sampling_rates_for_chunkid[sampling_rates_for_chunkid['ITEMID'] == 220179]
```

## Boxplot of Sampling Rates stratified by ITEMID 

As expected, we observe the lowest number of sampling rates for the NBP. The maximum for all three seem to be higher in the analysis for ICUSTAY_ID/ITEMID (as expected).

```python
# Imports
import seaborn as sns
import matplotlib.pyplot as plt

# Set variables
title = "Sampling Rate by ITEMID for Chunks"
xlabel = "ITEMID"
ylabel = "Average Number of Samples Obtained in one Hour"
plotdata = sampling_rates_for_chunkid
xvalue = "ITEMID"
yvalue = "SAMPLING_RATE"

# Config figure
sns.set_style("whitegrid")
fig, ax = plt.subplots(
    figsize = (10, 5), 
    dpi = 72 
    )
ax = sns.boxplot( 
    data = plotdata, 
    x = xvalue,
    y = yvalue, 
    palette = sns.color_palette("colorblind")
    )
ax.set_title(title, fontweight='bold', color= 'black', fontsize=14, y=1.05)
ax.set_xlabel(xlabel, fontsize=12, labelpad=15)
ax.set_ylabel(ylabel, fontsize=12, labelpad=15) 
ax.grid(b=True, which='both')
ax.margins(.1)

# Plot figure
plt.show(fig)
```
